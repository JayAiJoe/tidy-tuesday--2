<h1>Analysis of Video Game Sales</h1>

<h2>Libraries</h2>
Import necessary libraries for data handling and visualization.

```python
import os

#Data analysis and data prep
import pandas as pd
import numpy as np

from ipywidgets import interact

#Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2>Data Frames</h2>
Load the video game sales data.

```python
sales_df = pd.read_csv('vgsales.csv')
```

<h2>Data Preparation</h2>

First, check for missing values.

```python
sales_df.isnull().sum()
```

Only the Year and publisher columns have missing values. The rest of the columns have complete data. The missing Publisher data can be set to 'Unknown' so that the sales are still counted.  

```python
sales_df['Publisher'].fillna('Unknown', inplace = True)
sales_df.isnull().sum()
```

Rows with missing Year data can be dropped when analyzing yearly trends, but the other data in these rows should still be counted for anlyses that don't require the year. Hence, a separate dataframe should be made for yearly analyses.

```python
year_df = sales_df[sales_df['Year'].notna()]
year_df.isnull().sum()
```

<h2>Analysis and Visualizations</h2>

<h3>Publishers</h3>

The biggest and most popular game publishers can be identified based on the games that they publish and the sales that they make.

```python
pubs_count = sales_df['Publisher'].value_counts().nlargest(10)

#visualize
plt.figure(figsize=(15,8))
sns.barplot(x=pubs_count.values,y=pubs_count.index)
plt.ylabel('Publisher', fontsize=14)
plt.xlabel('Games Published', fontsize=14)
plt.title('Top 10 Publishers by Number of Games Published', fontsize=18)
plt.show()
```

```python
pubs_sales = sales_df.groupby(['Publisher'])['Global_Sales'].sum().nlargest(10)

#visualize
plt.figure(figsize=(15,8))
sns.barplot(x=pubs_sales.values,y=pubs_sales.index)
plt.ylabel('Publisher', fontsize=14)
plt.xlabel('Total Sales in Millions', fontsize=14)
plt.title('Top 10 Publishers by Global Sales', fontsize=18)
plt.show()
```

<h3>Regions</h3>

The regional sales reflect the popularity of video games in different parts of the world.

```python
NA_sales = year_df.groupby(['Year'])['NA_Sales'].sum()
EU_sales = year_df.groupby(['Year'])['EU_Sales'].sum()
JP_sales = year_df.groupby(['Year'])['JP_Sales'].sum()
Other_sales = year_df.groupby(['Year'])['Other_Sales'].sum()
```

```python
fig = plt.figure(figsize=(15,7))

NA_sales.plot(color='blue', label = 'North America')
EU_sales.plot(color='green', label = 'Europe')
JP_sales.plot(color='red', label = 'Japan')
Other_sales.plot(color='brown', label = 'Others')

#visualize
plt.xlabel('Year', fontsize=18)
plt.ylabel('Sales', fontsize=18)
plt.legend(loc ="upper left",fontsize=18)
plt.title('Regional Sales per Year', fontsize=22)
```

North America is the biggest consumer of video games followed by Europe then Japan. Japan, being only one country generally has higher video game sales than the other countries outside the big three. Video game sales have continued to rise from the 90's and has peaked around 2010. After 2010, video game sales have declined across all regions, though this may be due to the lack of updates in the database.

<h3>Genres</h3>


```python
genre_count = year_df.groupby(['Genre'])['Genre'].count().sort_values(ascending=False)
#visualize
plt.figure(figsize=(15,8))
sns.barplot(x=genre_count.values,y=genre_count.index)
plt.ylabel('Genre', fontsize=14)
plt.xlabel('Games Published', fontsize=14)
plt.title('Number of Published Games by Genre', fontsize=18)
plt.show()
```

```python
genre_sales = year_df.groupby(['Genre'])['Global_Sales'].sum().sort_values(ascending=False)
#visualize
plt.figure(figsize=(15,8))
sns.barplot(x=genre_sales.values,y=genre_sales.index)
plt.ylabel('Genre', fontsize=14)
plt.xlabel('Copies Sold in Millions', fontsize=14)
plt.title('Total Sales by Genre', fontsize=18)
plt.show()
```

The Action, Sports, RPG and Shooter genres seem to be the most popular genres for both the publishers and consumers. Generally, the relative number of games published corresponds to the the total sales in that genre. This shows the directly proportional relationship between games published and copies sold. Publishers will publish games that sell well and are in demand. 

The charts above also give insight on the distribution of gamer preferences. The mainstream genres are action, sports, RPGs and shooters while the more niche genres include fighting, simulation, puzzle and strategy games. 

```python
dfs = [x.groupby(['Genre'])['Genre'].count() for _, x in year_df.groupby('Year')]
genres = [[] for j in range(12)]
gcode = {'Action':0,'Sports':1,'Shooter':2,'Role-Playing':3,'Platform':4,'Misc':5,'Racing':6,'Fighting':7,'Simulation':8,'Puzzle':9,
        'Adventure':10,'Strategy':11}
genrelist = ['Action','Sports','Shooter','Role-Playing','Platform','Misc','Racing','Fighting','Simulation','Puzzle',
        'Adventure','Strategy']
years = [x for x in range(1980,2018)]
years.append(2020)

for x in dfs:
    ctr = 0
    for g in genrelist:
        if g in x.index:
            genres[gcode[g]].append(x[ctr])
            ctr += 1
        else:
            genres[gcode[g]].append(0)

fig = plt.figure(figsize=(15,7))

for g in genres:
    plt.plot(years,g,label=genrelist[genres.index(g)])

#visualize
plt.xlabel('Year', fontsize=18)
plt.ylabel('Games Published', fontsize=18)
plt.legend(loc ="upper left",fontsize=15)
plt.title('Genre Popularity Per Year', fontsize=22)
```

The popularity of video game genres could be due to multiple factors such as innovations in technology and game design and the consumer preferences of the time. Due to these factors, analyzing genre popularity through the years could give some valuable insight.

Action has been a consistently popular genre all since the 1980s. Video games surged in popularity and diversity sround 1995. The different genres experienced similar growth until around 2010. The RPG, Adenture and Sports genres experienced the most growth in this time. Most genres, except Action experienced a steep declined around 2012.

<h3>Platforms</h3>

The platforms define the technical limits of video games. Business pactices such as platform-exclusives also affect game sales and popularity. Here is a look into the progression of platforms through the years.

```python
platform = year_df.groupby(['Year','Platform'])['Global_Sales'].sum()

top5_yearly = [platform[y].nlargest(5) for y in range(1980,2018)]

def f(a):
    print('Top 5 Platforms of ' + str(a))
    return top5_yearly[a-1980].to_frame()

interact(f, a=(1980,1980,2017))
```

From the tables above, one can observe the popularity of the platforms is mainly driven by novelty and innovation. Nintendo was able to consistently stay on top because of their multiple platforms such as the NES, Gameboy and Wii.
